```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_results_2d_3d = pd.read_csv("results_2d_3d_all.csv")
df_results_2d_3d.head()
```

```python
df_original_result = df_results_2d_3d[(df_results_2d_3d.cube_dim == 3) & (df_results_2d_3d.space_dim == 100)]

fig, axs = plt.subplots(1, 2, figsize = (8, 4))
for ax, noise_coef in zip(axs, [0.2, 1]):
    df1 = df_original_result[df_original_result.noise_coef == noise_coef]
    ax.plot(df1.distortion, df1.sd, label = 'sd')
    ax.plot(df1.distortion, df1.ccgp, label = 'ccgp')
    ax.set_title(f"Noise coef = {noise_coef}")
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_ylim(0.5, 1.01)
    ax.set_xlabel("Distortion")
axs[0].legend()
```

```python
df_original_result = df_results_2d_3d[(df_results_2d_3d.cube_dim == 3)]

fig, axs = plt.subplots(2, 2, figsize = (10, 10))
for ax, noise_coef in zip(axs[0], [0.2, 1]):
    for distortion in [0, 0.5, 1, 1.5, 2]:
        df1 = df_original_result[(df_original_result.noise_coef == noise_coef)&(df_original_result.distortion == distortion)]
        ax.plot(df1.space_dim, df1.sd, label = f'sd, distortion = {distortion}')
        # ax.plot(df1.distortion, df1.ccgp, label = 'ccgp')
    ax.set_title(f"Noise coef = {noise_coef}")
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_ylim(0.5, 1.01)
    ax.set_xlabel("Embedded space dimension")
axs[0][0].legend()

for ax, noise_coef in zip(axs[1], [0.2, 1]):
    for distortion in [0, 0.5, 1, 1.5, 2]:
        df1 = df_original_result[(df_original_result.noise_coef == noise_coef)&(df_original_result.distortion == distortion)]
        ax.plot(df1.space_dim, df1.ccgp, label = f'ccgp, distortion = {distortion}')
        # ax.plot(df1.distortion, df1.ccgp, label = 'ccgp')
    ax.set_title(f"Noise coef = {noise_coef}")
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_ylim(0.5, 1.01)
    ax.set_xlabel("Embedded space dimension")
axs[0][0].legend()
axs[1][0].legend()

```

```python
# continuous data
import pandas as pd
with open("../cont_linear_ccgp.log", 'r') as f:
    lines = f.readlines()
    records = []
    for line in lines:
        record = line.split("ccgp,")[1].strip().split(',')
        embedding_space, num_variable, noise_coef, mode, score, coef = \
            int(record[0]), int(record[1]), float(record[2]), record[3], \
            float(record[4]), float(record[5])
        records.append([embedding_space, num_variable, noise_coef, mode, score, coef])
    df_ccgp = pd.DataFrame(records, columns=
        ['embedding_space', 'num_variable', 'noise_coef', 'mode', 'score', 'coef'])
df_ccgp.head()
```

```python
df_all_results = df_ccgp[df_ccgp['mode'] == 'normal']
fig, axs = plt.subplots(1, 4, figsize = (10, 3), sharex=True)
for ax, num_variable in zip(axs, [2, 3, 4, 5]):
    for noise_coef in [0, 0.2, 0.6, 1]:
        df1 = df_all_results[(df_all_results.noise_coef == noise_coef) & (df_all_results.num_variable == num_variable)]
        ax.plot(df1.embedding_space, df1.coef, label = f'ccgp, noise_coef={noise_coef}')
        # ax.plot(df1.distortion, df1.ccgp, label = 'ccgp')
        ax.set_title(f"Normal, #Var = {num_variable}")
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    # ax.set_ylim(0.5, 1.01)
    ax.set_xlabel("Embedded space dimension")
axs[-1].legend()
axs[0].set_ylabel("Corr. Coef.")
```

```python
df_all_results = df_ccgp[df_ccgp['mode'] == 'generalization']
fig, axs = plt.subplots(1, 4, figsize = (10, 3), sharex=True)
for ax, num_variable in zip(axs, [2, 3, 4, 5]):
    for noise_coef in [0, 0.2, 0.6, 1]:
        df1 = df_all_results[(df_all_results.noise_coef == noise_coef) & (df_all_results.num_variable == num_variable)]
        ax.plot(df1.embedding_space, df1.coef, label = f'regression metric, noise_coef={noise_coef}')
        # ax.plot(df1.distortion, df1.ccgp, label = 'ccgp')
        ax.set_title(f"Abstract, #Var = {num_variable}")
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    # ax.set_ylim(0.5, 1.01)
    ax.set_xlabel("Embedded space dimension")
axs[-1].legend()
axs[0].set_ylabel("Corr. Coef.")
# axs[1][0].legend()
```
