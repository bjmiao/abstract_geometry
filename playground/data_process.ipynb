```python
# continuous data
import pandas as pd
with open("../cont_linear_ccgp.log", 'r') as f:
    lines = f.readlines()
    records = []
    for line in lines:
        record = line.split("ccgp,")[1].strip().split(',')
        embedding_space, num_variable, noise_coef, mode, score, coef = \
            int(record[0]), int(record[1]), float(record[2]), record[3], \
            float(record[4]), float(record[5])
        records.append([embedding_space, num_variable, noise_coef, mode, score, coef])
    df_ccgp = pd.DataFrame(records, columns=
        ['embedding_space', 'num_variable', 'noise_coef', 'mode', 'score', 'coef'])
df_ccgp.head()
```

```python
import seaborn as seaborn
import matplotlib.pyplot as plt
import numpy as np

x1 = df_nsample_effect.embedding_space.unique()
y1 = df_nsample_effect.nsample.unique()
# z1 = df_nsample_effect.sd

def FUNC_Z(X1, Y1):
    Z1 = np.empty_like(X1, dtype='float')

    for i in range(X1.shape[0]):
        for j in range(X1.shape[1]):
            emb = X1[i, j]
            nsample = Y1[i, j]
            Z1[i, j] = float(df_nsample_effect[
                (df_nsample_effect.embedding_space == emb)
              & (df_nsample_effect.nsample == nsample)].ccgp)
    return Z1
X1, Y1 = np.meshgrid(x1, y1)
Z1 = FUNC_Z(X1, Y1)

plt.figure(figsize=(10, 10))
axes = plt.axes(projection="3d")
axes.plot_wireframe(X1, Y1, Z1)

axes.plot(np.arange(600), np.arange(600), 0.7, lw=2, ls='--', color='black')
axes.plot([600] * 400, np.arange(600, 1000), 0.7, lw=2, ls='--', color='black')
axes.plot(np.arange(0, 600), [1000] * 600, 0.7, lw=2, ls='--', color='black')
axes.plot([0] * 1000, np.arange(0, 1000), 0.7, lw=2, ls='--', color='black')

axes.set_xlabel("Embedding Space")
axes.set_ylabel("#Samples")
axes.set_zlabel("CCGP")
plt.show()
```
